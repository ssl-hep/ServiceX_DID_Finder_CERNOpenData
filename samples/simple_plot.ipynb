```python
from func_adl_servicex import ServiceXSourceUpROOT
from servicex import ServiceXDataset
import matplotlib.pyplot as plt
import awkward as ak
```

```python
sx_dataset = ServiceXDataset("cernopendata://3827", backend_type='dev_uproot')
ds = ServiceXSourceUpROOT(sx_dataset, "mini")
```

```python
data = ds.Select("lambda e: {'JetPT': e['jet_pt']}") \
    .AsAwkwardArray() \
    .value()
```

```python
plt.hist(ak.flatten(data["JetPT"]/1000.0), bins=40, range=(0.0, 400))
plt.title('Jet $p_T$')
_ = plt.xlabel('$p_T$ [GeV]')
```
